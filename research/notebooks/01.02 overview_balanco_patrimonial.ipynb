```python
from pipelines.readers.cvm_balanco_patrimonial import ReaderSQLCVMBalancoPatrimonial

import pandas as pd
import plotly.express as px
```

```python
reader = ReaderSQLCVMBalancoPatrimonial()

df = reader.read(
    ticker="VALE3",
    tipo_arquivo="BPA_con",
    cd_conta="1"
)

df
```

```python
df_plot = df.copy()
df_plot["Date"] = pd.to_datetime(df_plot["Date"])
value_col = df_plot.columns[1]
df_plot = df_plot.sort_values("Date")

fig = px.bar(
    df_plot,
    x="Date",
    y=df_plot[value_col].astype(float),
    title=f"Evolucao de {value_col}",
    template="plotly_white",
)

fig.update_layout(
    xaxis_title="Data",
    yaxis_title=value_col,
)

fig.show()
```
